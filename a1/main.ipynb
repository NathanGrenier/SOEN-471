```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# This code is just to make sure your environment is set up correctly. TODO: Should be deleted later in the project.

print(f"Pandas version: {pd.__version__}")
print(f"Numpy version: {np.__version__}")
print(f"Seaborn version: {sns.__version__}")

# Sample Data
np.random.seed(42)
data = pd.DataFrame(
    {
        "x": np.random.randn(100),
        "y": np.random.randn(100) + np.linspace(0, 5, 100),
        "category": np.random.choice(["Group A", "Group B"], 100),
    }
)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="x", y="y", hue="category", s=100)
plt.title("Environment Check: Plotting works!")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```
